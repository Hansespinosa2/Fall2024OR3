```python
import numpy as np

import matplotlib.pyplot as plt

# Define the constraint equation
def constraint(x):
	return 3 - 3*x

# Generate x values

x = np.linspace(0, 1, 100)

# Calculate the objective function values
y = x + constraint(x)

# Create the plot
plt.plot(x, y, label='Objective Function: max(x1 + x2)')
plt.plot(x, constraint(x), label='Constraint: 3x1 + x2 <= 3')
plt.fill_between(x, 0, constraint(x), where=(x >= 0) & (x <= 1) & (constraint(x) >= 0), alpha=0.3)

# Set plot labels and title
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Feasible Region and Objective Function')

# Set plot limits
plt.xlim(0, 1)
plt.ylim(0, 4)

# Add legend
plt.legend()

# Show the plot

plt.savefig('notes/images/08282024example.png')
plt.show()
```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

import matplotlib.pyplot as plt

# Define the function
def f(x1, x2):
	return 2 * x1**2 + 2 * x2**2

# Generate data points
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)
Y = f(X1, X2)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, Y, cmap='viridis', alpha= 0.75)
ax.plot([0], [0], [0], markerfacecolor='k', markeredgecolor='k', marker='o', markersize=5)

# Set plot labels and title
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f')

plt.savefig('notes/images/gradient_ex1.png')
# Show the plot
plt.show()
```

```python
import numpy as np

import matplotlib.pyplot as plt

# Define the function
def f(x):
	return (x-3)*(x**2-1)*(x+2)

# Generate x values
x = np.linspace(-5, 5, 100)

# Calculate y values
y = f(x)

# Create the plot
plt.plot(x, y, label='(x-3)(x^2-1)(x+2)')

# Plot the solutions
solutions = [2,0, -1.5]
f_solutions = [f(2),f(0), f(-1.5)]
plt.scatter(solutions, f_solutions, color='red', label='Solutions')

# Set plot labels and title
plt.xlabel('x')
plt.ylabel('y')

# Add legend
plt.legend()


plt.savefig('notes/images/gradient_ex4.png')
# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the vector df
df = np.array([1, 1])

# Generate the vector dg
dg_1 = np.array([-1, 0])
dg_2 = np.array([-1, -.25])



# Create a figure and axis
fig, ax = plt.subplots()
fig.set_figheight(5)
fig.set_figwidth(5)

# Color x>0 as very light blue
ax.fill_betweenx(np.arange(-2, 2, 0.01), 0, 2, color='lightblue', alpha=0.3)
x = np.linspace(-2,2,400)
y = -x
ax.fill_between(x,y, -2, color='lightcoral', alpha=0.3)

y_g_2 = -4*x
ax.fill_between(x,y_g_2, 2, color='lightgreen', alpha=0.3) 

ax.fill_between(x, y, y_g_2, where=(y_g_2 <= y), color='yellow', alpha=0.3)



# Plot the vectors
ax.quiver(0, 0, df[0], df[1], angles='xy', scale_units='xy', scale=1, color='r', label='df')
ax.quiver(0, 0, dg_1[0], dg_1[1], angles='xy', scale_units='xy', scale=1, color='b', label='dg_1')
ax.quiver(0, 0, dg_2[0], dg_2[1], angles='xy', scale_units='xy', scale=1, color='g', label='dg_2')


# Set plot limits
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])

# Set plot labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Vectors df and dg')

# Add legend
ax.legend()

plt.savefig('notes/images/walkable_region.png')

# Show the plot
plt.show()
```
