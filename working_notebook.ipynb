```python
import numpy as np

import matplotlib.pyplot as plt

# Define the constraint equation
def constraint(x):
	return 3 - 3*x

# Generate x values

x = np.linspace(0, 1, 100)

# Calculate the objective function values
y = x + constraint(x)

# Create the plot
plt.plot(x, y, label='Objective Function: max(x1 + x2)')
plt.plot(x, constraint(x), label='Constraint: 3x1 + x2 <= 3')
plt.fill_between(x, 0, constraint(x), where=(x >= 0) & (x <= 1) & (constraint(x) >= 0), alpha=0.3)

# Set plot labels and title
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Feasible Region and Objective Function')

# Set plot limits
plt.xlim(0, 1)
plt.ylim(0, 4)

# Add legend
plt.legend()

# Show the plot

plt.savefig('notes/images/08282024example.png')
plt.show()
```
